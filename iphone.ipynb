## How many rating do the highest rated iphone on flipkart have ?

```python
df.head(3)
```

```python
df.sort_values(by='Star Rating', ascending=False).iloc[0][['Number Of Ratings', 'Product Name']]
```

```python
# Find the highest-rated iPhone
highest_rated_iphone = df.sort_values(by='Star Rating', ascending=False).iloc[0]

# Get the number of ratings
number_of_ratings = highest_rated_iphone['Number Of Ratings']
print(f"The highest-rated iPhone has {number_of_ratings} ratings.")
```

```python
import matplotlib.pyplot as plt
top_10_iphones = df.sort_values(by='Star Rating', ascending=False).head(10)

# Plot the bar graph
plt.figure(figsize=(12, 10))
plt.bar(top_10_iphones['Product Name'], top_10_iphones['Number Of Ratings'], color='skyblue')
plt.xlabel('Product Name')
plt.ylabel('Number Of Ratings')
plt.title('Top 10 Highest Rated iPhones on Flipkart')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

## What is the relationship between the sale price of iphone and the number of rating on flipkart?

```python
df.head(2)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Sale Price'], df['Number Of Ratings'], color='blue', alpha=0.5)
plt.xlabel('Sale Price (INR)')
plt.ylabel('Number Of Ratings')
plt.title('Relationship between Sale Price and Number of Ratings for iPhones on Flipkart')
plt.grid(True)
plt.show()
```

## What is the relationship between the discount percentage and the number of ratings of iphone on flipkart?

```python
df.head(3)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df['Discount Percentage'], df['Number Of Ratings'], color='blue', alpha=0.5)
plt.xlabel('Discount Percentage (INR)')
plt.ylabel('Number Of Ratings')
plt.title('Relationship between Discount percentage and Number of Ratings for iPhones on Flipkart')
plt.grid(True)
plt.show()
```
